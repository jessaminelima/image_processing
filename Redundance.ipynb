A sequência de quadros está no arquivo denominado frame_mpeg.mat.
Exibir a sequência de quadros recuperada para se obter um vídeo (4,0)

```python
from scipy.io import loadmat
from PIL import Image as im
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct, idct
import os
```

```python
img = loadmat("frame_mpeg.mat") # leitura do arquivo
```

```python
blck = np.array(img["blocks"])
print(blck.shape)
```

```python
frame = []
for i in range (252):
    frame.append(blck[0][i])
```

```python
print("*****Original*****\n")
fig,aux = plt.subplots(1, 5, figsize = (18, 15))

for i in range(5):
   aux[i].imshow(frame[i])
   aux[i].set_title(f" Bloco {i+1}")
```

```python
for i in range(0, 250, 5):
    frame[i+1] = frame[i] - frame[i+1]
    frame[i+2] = frame[i] - frame[i+2]
    frame[i+3] = frame[i] - frame[i+3]
    frame[i+4] = frame[i] - frame[i+4]
```

```python
print("*****Reconstituido*****\n")
fig,aux = plt.subplots(1, 5, figsize = (18, 15))

for i in range(5):
   aux[i].imshow(frame[i])
   aux[i].set_title(f" Bloco {i+1}")
```
